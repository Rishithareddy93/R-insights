```r
# Load required libraries
library(dplyr)
library(ggplot2)
library(gganimate)
library(gifski)
library(transformr)
library(readr)
library(fmsb)
# Read CSVs
matches <- read_csv("matches.csv")
deliveries <- read_csv("deliveries.csv")
```

```r
matches %>%
  group_by(season) %>%
  summarise(total_matches = n()) %>%
  ggplot(aes(x = factor(season), y = total_matches)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  labs(title = "Total Matches per Season", x = "Season", y = "Matches")
```

```r
matches %>%
  filter(!is.na(winner)) %>%
  count(winner, sort = TRUE) %>%
  top_n(10) %>%
  ggplot(aes(x = reorder(winner, n), y = n)) +
  geom_bar(stat = "identity", fill = "darkgreen") +
  coord_flip() +
  labs(title = "Top 10 Teams by Wins", x = "Team", y = "Wins")
```

```r
# Join deliveries with match dates
deliveries_season <- deliveries %>%
  left_join(matches %>% select(id, date), by = c("match_id" = "id"))

# Filter top 5 batters overall
top_batters <- deliveries_season %>%
  group_by(batter) %>%
  summarise(total_runs = sum(batsman_runs)) %>%
  arrange(desc(total_runs)) %>%
  slice(1:5) %>%
  pull(batter)

# Create cumulative data over time
cumulative_runs <- deliveries_season %>%
  filter(batter %in% top_batters) %>%
  group_by(batter, date) %>%
  summarise(daily_runs = sum(batsman_runs), .groups = "drop") %>%
  arrange(batter, date) %>%
  group_by(batter) %>%
  mutate(cumulative_runs = cumsum(daily_runs))
```

```r
animated_plot <- ggplot(cumulative_runs, aes(x = date, y = cumulative_runs, color = batter)) +
  geom_line(size = 1.2) +
  labs(title = "Cumulative Runs by Top 5 Batters Over Time",
       x = "Date", y = "Cumulative Runs") +
  theme_minimal() +
  transition_reveal(date)

animated_plot
```

```r
library(dplyr)
library(fmsb)
library(tidyr)


# Total runs scored by each team
batting_stats <- deliveries %>%
  group_by(batting_team) %>%
  summarise(total_runs = sum(total_runs), .groups = "drop")

# Wickets taken by each team
bowling_stats <- deliveries %>%
  filter(!is.na(player_dismissed), dismissal_kind != "run out") %>%
  group_by(bowling_team) %>%
  summarise(wickets = n(), .groups = "drop")

# Toss wins by each team
toss_stats <- matches %>%
  group_by(toss_winner) %>%
  summarise(toss_wins = n(), .groups = "drop") %>%
  rename(team = toss_winner)

# Combine all metrics into one table
team_stats <- full_join(batting_stats, bowling_stats, by = c("batting_team" = "bowling_team")) %>%
  rename(team = batting_team) %>%
  full_join(toss_stats, by = "team") %>%
  replace_na(list(total_runs = 0, wickets = 0, toss_wins = 0))
```

```r
radar_data <- as.data.frame(team_stats)
rownames(radar_data) <- radar_data$team
radar_data$team <- NULL


# Add min/max rows
radar_data <- rbind(
  apply(radar_data, 2, max),
  apply(radar_data, 2, min),
  radar_data
)
```

```r
# 6. Choose a few teams to plot (e.g., top 4 by total_runs)
selected_teams <- rownames(radar_data)[3:6]  # Skip max and min rows

# 7. Plot the radar chart
radarchart(radar_data[c(1, 2, selected_teams), ],
           axistype = 1,
           pcol = rainbow(length(selected_teams)),
           plwd = 2,
           plty = 1,
           title = "IPL Team Performance Radar Chart (Runs, Wickets, Toss Wins)",
           cglcol = "grey", cglty = 1, axislabcol = "grey")

legend("topright", legend = selected_teams, col = rainbow(length(selected_teams)), lty = 1, lwd = 2)
```
